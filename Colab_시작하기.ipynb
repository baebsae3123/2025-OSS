```
import numpy as np

fake_a = 2
fake_b = 80

x = np.array([2, 4, 6, 8])
y = np.array([81, 93, 91, 97])

def predict(x, a, b):
    return a * x + b

def mse(y, y_pred):
    n = len(y)
    return (1/n) * sum((y - y_pred) ** 2)

# fake_a를 2씩 올리면서 5번 반복
for step in range(20):
    predict_result = []  # 매 반복마다 새로 생성

    for i in range(len(x)):
        y_pred = predict(x[i], fake_a, fake_b)
        predict_result.append(y_pred)
        print("이동거리=%.f km, 실제점수=%.f, 예측점수=%.f" % (x[i], y[i], y_pred))

    print("평균 제곱오차:", mse(y, predict_result))
    print("\n")  # 줄바꿈
    fake_a += 0.1  # 다음 반복에 a 증가
    fake_b -= 0.1
```

```
import numpy as np
import matplotlib.pyplot as plt

x= np.array([2,4,6,8])
y = np.array([81,93,91,97])

a = 0
b = 0
#학습률 정합니다.
lr = 0.03

# 몇번 반복될지 설정
epochs = 10000

# x 값이 총 몇개인지 셉니다.
n = len(x)

for i in range(epochs):

    y_pred = a * x + b
    error = y  - y_pred

    a_diff = (2/n) * sum(-x * (error))
    b_diff = (2/n) * sum (-(error))

    a = a- lr * a_diff
    b = b- lr * b_diff

    if i % 100 == 0:
      print("epoch =%.f, 기울기 =%.04f , 절편 %.04f" % (i,a,b))
```
