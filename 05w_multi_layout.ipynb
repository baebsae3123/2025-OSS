```python
fig, ax = plt.subplots(2, 2)

x = np.arange(5)

ax[0, 0].plot(x,x)
ax[0, 0].set_title("Y=x")

ax[0, 1].plot(x,2*x)
ax[0, 1].set_title("Y=2x")

ax[1, 0].plot(x,x**2)
ax[1, 0].set_title("Y=x**2")

ax[1, 1].plot(x,x**3)
ax[1, 1].set_title("Y=x**3")

fig.suptitle("subplot 그래프",fontsize=16)
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(2, 2)

x = np.arange(5)

ax[0][0].plot(x,x)
ax[0][0].set_title("Y=x")

ax[0][1].plot(x,2*x)
ax[0][1].set_title("Y=2x")

ax[1][0].plot(x,x**2)
ax[1][0].set_title("Y=x**2")

ax[1][1].plot(x,x**3)
ax[1][1].set_title("Y=x**3")

fig.suptitle("subplot 그래프",fontsize=16)
plt.tight_layout()
plt.show()
```

```python
fig ,ax = plt.subplots(2,2)

ax[0,0].plot([1,2,3],[1,2,3])
ax[0,0].set_title("선그래프",fontsize=8)
ax[0,1].bar([1,2,3],[3,2,1])
ax[0,1].set_title("막대 그래프",fontsize=8)
ax[1,0].scatter([1,2,3,],[3,2,1])
ax[1,0].set_title("산점도그래프",fontsize=8)
ax[1,1].hist([1,2,2,3,3],bins=[.5,1.5,2.5,3.5],rwidth=.95)
ax[1,1].set_title("돗수분포 그래프",fontsize=8)

fig.suptitle("전체 그래프",fontsize=12)
plt.tight_layout()
plt.show()
```

```python
fig ,ax = plt.subplots(2,2)

for i , axi in enumerate(ax.flat):
  axi.plot([1,2,3],[n*(i+1) for n in [1,2,3]])
  axi.set_title(f"Plot {i+1}")

plt.tight_layout()
plt.show()
```
